```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import sys
import PIL
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img

DIRDATA = "./todaslascaras/original"
SOLOCARAS = "./todaslascaras/SOLOCARAS"
TODASLASCARAS = "./todaslascaras"
PREPRO = "./todaslascaras/prepro-2.0"
#CATEGORIES = ["anger", "disgust"]
CATEGORIES = ["anger", "fear", "happy", "joy", "kiss", "neutral", "sadness", "surprise"]
ANGER = ["anger"]
OBJECTIVEDATA = "./todaslascaras/prepro-2.0"

```

PRIMERO OBTENGO EL RECUADRO DE LA CARA Y LO GUARDO EN UN DIRECTORIO, SEPARANDO POR CLASES

```python
def detectar_y_guardar_cara(path, savePath, category):
    count = 0
    for img in os.listdir(path):
        count += 1
        image = cv2.imread(path+'/'+img)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
        faces = faceCascade.detectMultiScale(
            gray,
            scaleFactor=1.03,
            minNeighbors=6,
            minSize=(30, 30)
        )
        for (x, y, w, h) in faces:
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
            roi_color = image[y:y + h, x:x + w]
            cv2.imwrite(savePath + '/' + category +  str(w) + str(h) + '_faces.jpg', roi_color)
```

```python
def generar_imagenes(where, categories):
    for category in categories:
        path = os.path.join(DIRDATA, category)
        detectar_y_guardar_cara(path, where+'/'+category, category)
```

```python
generar_imagenes(where='todaslascaras/SOLOCARAS', categories=CATEGORIES)
```

A CONTINUACION SE ESPEJAN TODAS LAS CARAS Y SE GUARDAN EN EL DIRECTORIO 'SOLOCARAS'

```python
#Con esto leo la imagen y la devuelvo en el color que me indican
def leer_imagen(img_path, gray_scale=False):
    image_src = cv2.imread(img_path)
    if gray_scale:
        image_rgb = cv2.cvtColor(image_src, cv2.COLOR_BGR2GRAY)
    else:
        image_rgb = cv2.cvtColor(image_src, cv2.COLOR_BGR2RGB)
    return image_rgb

#Este metodo espeja una imagen que viene por parametro y la devuelve
def espejar_imagen(img_path, gris=False):
    img = leer_imagen(img_path, gris)
    img_mirrored = np.fliplr(img)
    return img_mirrored
```

```python
def guardar_imagen(img, name, where):
    cv2.imwrite(where + '/' + name, img)

def espejar_todo(where, categories):
    for category in categories:
        path = os.path.join(SOLOCARAS, category)
        for img in os.listdir(path):
            mirrored_img = espejar_imagen(path+'/'+img, gris = True)
            guardar_imagen(mirrored_img, img + '_mirrored.jpg', where = where +'/'+category)
```

```python
espejar_todo(where=SOLOCARAS,categories=CATEGORIES)
```

CON LAS CARAS ESPEJADAS, TOMO CADA IMAGEN Y OBTENGO 10 COPIAS ROTADAS EN UN GRADO DE 22

```python
rotate_gen = ImageDataGenerator(rotation_range=22)
def rotar_y_guardar(cantidadImagenes, path, category):
    for img in os.listdir(path):
            i = load_img(path+'/'+img, color_mode="grayscale", target_size=(120,120))
            i = img_to_array(i)
            i = np.expand_dims(i, axis=0)
            PREFIX = category
            imGen = rotate_gen.flow(i, batch_size=1, save_to_dir=PREPRO+'/'+category, save_prefix=PREFIX, save_format='jpg')
            for x in range(cantidadImagenes):
                batch = imGen.next()
```

```python
def rotar_caras():
    for category in CATEGORIES:
        path = os.path.join(SOLOCARAS, category)
        rotar_y_guardar(10, SOLOCARAS+'/'+category, category)
```
